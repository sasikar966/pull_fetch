```python
x = np.array([8, 5, 7, 9, 10, 12, 11,  25])
y = np.array([4, 5, 6, 7, 8, 10, 20, 30])
plt.plot(x, y)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data")
plt.show()
```

```python

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.show()
```

```python
x = np.array(["a", "b", "c", "d"])
y = np.array([2, 6, 2, 8])
plt.bar(x,y, width = 0.5)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 6, 8, 10])
plt.barh(x, y, color = "red", height = 0.5)
plt.show()
```

```python
y = np.array([25, 45, 35, 65])
plt.pie(y)
plt.show()
```

```python
y = np.array([15, 35, 45, 25])
mylabels = ["car", "bike", "train", "bus"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```
